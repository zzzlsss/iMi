```python
ls = hv.link_selections.instance(link_inputs=True)
sel_stream1 = hv.streams.Selection1D().rename(index='index1')
sel_stream2 = hv.streams.Selection1D().rename(index='index2')


def plot_h2_vs_h2o(index1, index2):
    indices = index1 if index1 and len(index1) > 0 else index2 if index2 and len(index2) > 0 else []
    if indices:
        selected = cat.iloc[indices]
        scatter = hv.Points(
            selected,
            kdims=['H2_N', 'H2O_N'], vdims=['ID', 'H2_N_sci', 'H2O_N_sci']
        ).opts(
            color='blue', size=6, marker='circle', alpha=0.7,
            tools=['hover'], nonselection_alpha=0.7,
            xlabel='H2_N', ylabel='H2O_N', title='N H2 vs. N H2O',
            hover_tooltips=[
                ('ID', '@ID'),
                ('H2_N', '@H2_N_sci'),
                ('H2O_N', '@H2O_N_sci')
            ],
        )
    else:
        scatter = hv.Points(
            cat,
            kdims=['H2_N', 'H2O_N'], vdims=['ID', 'H2_N_sci', 'H2O_N_sci']
        ).opts(
            color='blue', size=6, marker='circle', alpha=0.4,
            tools=['hover'],
            xlabel='H2_N', ylabel='H2O_N', title='N H2 vs. N H2O',
            hover_tooltips=[
                ('ID', '@ID'),
                ('H2_N', '@H2_N_sci'),
                ('H2O_N', '@H2O_N_sci')
            ],
        )
    
    return scatter

scatter_H2O = hv.DynamicMap(plot_h2_vs_h2o, streams=[sel_stream1, sel_stream2]).opts(tools=['hover', 'tap', 'lasso_select'])

def plot_h2_vs_co2(index1,index2):
    indices = index1 if index1 and len(index1) > 0 else index2 if index2 and len(index2) > 0 else []
    if indices:
        selected = cat.iloc[indices]
        scatter = hv.Points(
            selected,
            kdims=['H2_N', 'CO2_N'], vdims=['ID','H2_N_sci', 'CO2_N_sci']
        ).opts(
            color='green', size=6, marker='circle', alpha=0.7,
            tools=['hover', 'tap', 'lasso_select'],
            nonselection_alpha=0.7,
            xlabel='H2_N', ylabel='CO2_N', title='N H2 vs. N CO2',
            hover_tooltips=[('ID', '@ID'), ('H2_N', '@H2_N_sci'), ('CO2_N', '@CO2_N_sci')],
        )
    else:
        scatter = hv.Points(
            cat,
            kdims=['H2_N', 'CO2_N'], vdims=['ID','H2_N_sci', 'CO2_N_sci']
        ).opts(
            color='green', size=6, marker='circle', alpha=0.4,
            tools=['hover', 'tap', 'lasso_select'],
            xlabel='H2_N', ylabel='CO2_N', title='N H2 vs. N CO2',
            hover_tooltips=[('ID', '@ID'), ('H2_N', '@H2_N_sci'), ('CO2_N', '@CO2_N_sci')],
        )
    return scatter

scatter_CO2 = hv.DynamicMap(plot_h2_vs_co2, streams=[sel_stream1, sel_stream2]).opts(tools=['hover', 'tap', 'lasso_select'])

img = ls(scatter_H2O + scatter_CO2)

img
```
